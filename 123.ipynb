```python
# Import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Simple data for the graphs
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y1 = np.sin(x)              # Sine wave
y2 = np.cos(x)              # Cosine wave

# Line Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='Sine Wave', color='blue')
plt.plot(x, y2, label='Cosine Wave', color='orange')
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()

# Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(x, y1, label='Sine Points', color='red', alpha=0.6)
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()

# Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='green', alpha=0.7)
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```
